```python
from sklearn.feature_extraction.text import TfidfVectorizer

corpus = [
    "I love machine learning",
    "Machine learning is amazing",
    "I love coding and machine learning"
]

vectorizer = TfidfVectorizer()

X = vectorizer.fit_transform(corpus)

print("Feature Names (Words):\n")
print(vectorizer.get_feature_names_out())
print("\nTF-IDF Matrix:\n")
print(X.toarray())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

corpus = [
    "I love machine learning",
    "Machine learning is amazing",
    "I love coding and machine learning"
]

vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(corpus)

cos_sim = cosine_similarity(tfidf_matrix)

print("Cosine Similarity Matrix:\n")
print(cos_sim)
```

```python
import nltk
nltk.download('punkt_tab') # Changed from 'punkt' as per error message
from nltk.tokenize import word_tokenize
from collections import Counter
import math

corpus = [
    "I love machine learning",
    "Machine learning is amazing",
    "I love coding and machine learning"
]

tokens = []
for sentence in corpus:
    tokens.extend(word_tokenize(sentence.lower()))

word_counts = Counter(tokens)
total_words = len(tokens)

bigrams = list(nltk.bigrams(tokens))
bigram_counts = Counter(bigrams)
total_bigrams = len(bigrams)

def compute_pmi(w1, w2):
    p_w1 = word_counts[w1] / total_words
    p_w2 = word_counts[w2] / total_words
    p_w1_w2 = bigram_counts[(w1, w2)] / total_bigrams if (w1, w2) in bigram_counts else 0

    if p_w1_w2 == 0:
        return 0

    return math.log2(p_w1_w2 / (p_w1 * p_w2))

print("PMI Scores:\n")

vocab = list(word_counts.keys())

for i in range(len(vocab)):
    for j in range(i+1, len(vocab)):
        w1, w2 = vocab[i], vocab[j]
        pmi = compute_pmi(w1, w2)

        if pmi > 0:
            print(f"{w1} - {w2} : {round(pmi, 3)}")
```
