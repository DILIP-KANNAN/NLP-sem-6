```python
import nltk
from nltk.corpus import brown
from nltk.tokenize import word_tokenize

# Download once
nltk.download('brown')
nltk.download('punkt')
```

```python
nltk.download('punkt_tab')
# Sample text
text = "The teachers are jumping over fences."
tokens = word_tokenize(text)
```

```python
# Rule sets
prefixes = ['un', 're', 'in', 'dis', 'pre']
suffixes = ['ing', 'ed', 'ly', 'ness', 'ment', 's']
```

```python
# Morphological Analyzer
def analyze_word(word):
    root = word
    found_prefix = ''
    found_suffix = ''

    # check prefix
    for p in prefixes:
        if word.startswith(p) and len(word) > len(p) + 2:
            root = word[len(p):]
            found_prefix = p
            break

    # check suffix
    for s in suffixes:
        if root.endswith(s) and len(root) > len(s) + 2:
            root = root[:-len(s)]
            found_suffix = s
            break

    return {
        "word": word,
        "root": root,
        "prefix": found_prefix,
        "suffix": found_suffix
    }
```

```python
# Apply on sample tokens
results = [analyze_word(w) for w in tokens[:20]]

for r in results:
    print(r)
```

```python
# Finite State Automaton for suffix
def fsa_suffix(word):
    # States for detecting 'ing'
    if len(word) >= 3 and word.endswith("ing"):
        return word[:-3], "ing"

    # States for detecting 'ed'
    if len(word) >= 2 and word.endswith("ed"):
        return word[:-2], "ed"

    # States for detecting 'ly'
    if len(word) >= 2 and word.endswith("ly"):
        return word[:-2], "ly"

    # default
    return word, ""
```

```python
# Apply FSA analyzer
results = []

for w in tokens[:20]:
    root, suffix = fsa_suffix(w)
    results.append({
        "word": w,
        "root": root,
        "suffix": suffix
    })

for r in results:
    print(r)
```
