```python
import nltk
from nltk import word_tokenize, pos_tag, ne_chunk

# Download necessary NLTK data
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger_eng')
nltk.download('maxent_ne_chunker_tab')
nltk.download('words') # Added to resolve the LookupError for 'words'

# ===== SMALL DATASET =====
text = """
Sundar Pichai is the CEO of Google. He was born in India and works in California.
Microsoft Corporation is based in the United States.
"""

def process_text(data):
    tokens = word_tokenize(data)
    pos_tags = pos_tag(tokens)
    ner_tree = ne_chunk(pos_tags)

    results = []
    for subtree in ner_tree:
        if hasattr(subtree, 'label'):
            entity = " ".join([word for word, tag in subtree.leaves()])
            results.append(f"{entity} --> {subtree.label()}")
    return results

print("\nProcessing Small Dataset...\n")
output = process_text(text)

# ===== OUTPUT =====
print("Named Entities:\n")
for item in output:
    print(item)
```
