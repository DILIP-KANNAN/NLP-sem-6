```python
import nltk
from nltk.corpus import brown, reuters
from collections import Counter

# Download once
nltk.download('brown')
nltk.download('reuters')
```

```python
train_words = [w.lower() for w in brown.words()[:5000]]

# Count frequencies
word_counts = Counter(train_words)

N = len(train_words)                # total words
V = len(word_counts)                # vocabulary size

print("Total Words:", N)
print("Vocabulary Size:", V)
```

```python
def unigram_prob(word):
    return word_counts[word] / N if word in word_counts else 0

def laplace_prob(word):
    return (word_counts[word] + 1) / (N + V)
```

```python
test_words = [w.lower() for w in reuters.words()[:10]]
print(test_words)
print("\nWord Probabilities:\n")

for w in test_words:
    print(
        w,
        "Unigram:", round(unigram_prob(w),6),
        "Laplace:", round(laplace_prob(w),6),
        '\n'
    )
```
