```python
import nltk
import math
from collections import Counter
from nltk.corpus import brown, reuters

# Downloads (run once)
nltk.download('brown')
nltk.download('reuters')
```

```python
# Load dataset (small sample)
words = [w.lower() for w in brown.words()[:5000]]
```

```python
# Build unigram distribution
freq = Counter(words)
total_words = sum(freq.values())

prob = {w: c/total_words for w, c in freq.items()}
```

```python
# 1. Entropy
entropy = -sum(p * math.log2(p) for p in prob.values())
print("Entropy:", round(entropy, 4))
```

```python
# Cross-Entropy & Perplexity
# Using Reuters as test set

test_words = [w.lower() for w in reuters.words()[:1000]]
```

```python
# smoothing to avoid log(0)
epsilon = 1e-6

cross_entropy = 0
N = len(test_words)

for w in test_words:
    q = prob.get(w, epsilon)
    cross_entropy += -math.log2(q)

cross_entropy /= N

perplexity = 2 ** cross_entropy

print("Cross Entropy:", round(cross_entropy, 4))
print("Perplexity:", round(perplexity, 4))
```
