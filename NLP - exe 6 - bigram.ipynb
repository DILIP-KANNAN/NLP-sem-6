```python
import nltk
from nltk.corpus import brown, reuters
from collections import Counter

nltk.download('brown')
nltk.download('reuters')
```

```python
words = [w.lower() for w in brown.words()[:5000]]

# build bigrams
bigrams = list(nltk.bigrams(words))

bigram_counts = Counter(bigrams)
unigram_counts = Counter(words)
```

```python
count_of_counts = Counter(bigram_counts.values())

def good_turing(c):
    Nc = count_of_counts[c]
    Nc1 = count_of_counts.get(c+1, 1)
    return (c+1) * (Nc1 / Nc)
```

```python
def bigram_prob(w1, w2):
    c = bigram_counts.get((w1, w2), 0)

    if c > 0:
        c_star = good_turing(c)
    else:
        c_star = 1   # unseen event estimate

    return c_star / unigram_counts[w1] if unigram_counts[w1] else 0
```

```python
test_words = [w.lower() for w in reuters.words()[:10]]

print("Test words: ", test_words,"\n")
print("Bigram Probabilities:\n")

for i in range(len(test_words)-1):
    w1 = test_words[i]
    w2 = test_words[i+1]
    print((w1, w2), ":", round(bigram_prob(w1, w2),6))
```
